```python
import matplotlib.pyplot as plt

# í•¨ìˆ˜ ì •ì˜: ê¸ì •ê³¼ ë¶€ì •ì˜ ë¹„ìœ¨ ê³„ì‚° ë° ì‚°ì ë„ ê·¸ë¦¬ê¸°
def plot_sentiment_ratios(df, name):
    positive_ratios = df[df['ì˜ˆì¸¡ ê°ì •'] == "ê¸ì •"]['ëŒ“ê¸€ ê¸¸ì´'].value_counts() / df['ëŒ“ê¸€ ê¸¸ì´'].value_counts()
    negative_ratios = df[df['ì˜ˆì¸¡ ê°ì •'] == "ë¶€ì •"]['ëŒ“ê¸€ ê¸¸ì´'].value_counts() / df['ëŒ“ê¸€ ê¸¸ì´'].value_counts()

    plt.figure(figsize=(6, 4))
    plt.scatter(positive_ratios.index, positive_ratios, label='ê¸ì •', alpha=0.5)
    plt.scatter(negative_ratios.index, negative_ratios, label='ë¶€ì •', alpha=0.5)
    plt.xlabel('ëŒ“ê¸€ ê¸¸ì´')
    plt.ylabel('ê¸ì • ë° ë¶€ì • ë¹„ìœ¨')
    plt.legend()
    plt.title(f'{name}ì˜ ëŒ“ê¸€ ê¸¸ì´ì— ë”°ë¥¸ ê¸ì •ê³¼ ë¶€ì • ë¹„ìœ¨')
    plt.show()


# df_jinì— ëŒ€í•œ ì‚°ì ë„
plot_sentiment_ratios(df_jin, 'ì§„êµí›ˆ')

# df_kimì— ëŒ€í•œ ì‚°ì ë„
plot_sentiment_ratios(df_kim, 'ê¹€íƒœìš°')

# df_kwonì— ëŒ€í•œ ì‚°ì ë„
plot_sentiment_ratios(df_kwon, 'ê¶Œìˆ˜ì •')
```
