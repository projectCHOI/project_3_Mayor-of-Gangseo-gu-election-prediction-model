```python
df_jin = pd.read_csv("/content/drive/MyDrive/SeSAC_AI 전문인력 양성/Project/Raw Data/comment_진교훈.csv", header=None)
df_kim = pd.read_csv("/content/drive/MyDrive/SeSAC_AI 전문인력 양성/Project/Raw Data/comment_김태우.csv", header=None)
df_kwon = pd.read_csv("/content/drive/MyDrive/SeSAC_AI 전문인력 양성/Project/Raw Data/comment_권수정.csv", header=None)

print(df_jin.describe())
print(df_kim.describe())
print(df_kwon.describe())
```

```python
df_jin.columns = ["comment"]
df_kim.columns = ["comment"]
df_kwon.columns = ["comment"]
```

```python
# 'comment' 열의 각 행의 문자열 길이를 계산하여 'Comment_length' 열에 저장
df_jin['comment_length'] = df_jin['comment'].apply(lambda x: len(str(x)))
df_kim['comment_length'] = df_jin['comment'].apply(lambda x: len(str(x)))
df_kwon['comment_length'] = df_jin['comment'].apply(lambda x: len(str(x)))
```

```python
print(df_jin.describe())
print(df_kim.describe())
print(df_kwon.describe())
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
# 'comment_length' 열을 기준으로 정렬하여 상위 10개 행 출력
sorted_df_jin = df_jin.nlargest(80, 'comment_length', keep='all')

print(sorted_df_jin)
```

```python
selected_value = df_jin.loc[[11782, 1736, 862], 'comment']  # 'A' 열의 인덱스 2의 값 선택
selected_value
```

```python
duplicate_rows_jin = df_jin['comment'].duplicated().sum()
duplicate_rows_kim = df_kim['comment'].duplicated().sum()
duplicate_rows_kwon = df_kwon['comment'].duplicated().sum()

print(f'jin 중복된 행의 개수: {duplicate_rows_jin}')
print(f'kim 중복된 행의 개수: {duplicate_rows_kim}')
print(f'kwon 중복된 행의 개수: {duplicate_rows_kwon}')
```

```python
duplicate_rows = df_jin[df_jin.duplicated()]

duplicate_rows_sorted = duplicate_rows.sort_values(by=list(duplicate_rows.columns), ascending=False)

print(duplicate_rows_sorted)
```
