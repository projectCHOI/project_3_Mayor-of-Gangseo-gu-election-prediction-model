```python
# YouTube API í‚¤ë¥¼ ì„¤ì •í•©ë‹ˆë‹¤
api_service_name = "youtube"
api_version = "v3"
api_key = "AIzaSyAmPf8J_EzDXtAhRqJPJMLXDckWxzAeXJQ"  # ì—¬ê¸°ì— ë³¸ì¸ì˜ API í‚¤ë¥¼ ë„£ì–´ì£¼ì„¸ìš”

# YouTube API í´ë¼ì´ì–¸íŠ¸ë¥¼ ìƒì„±í•©ë‹ˆë‹¤
youtube = googleapiclient.discovery.build(
    api_service_name, api_version, developerKey=api_key)

# ê²€ìƒ‰ì–´ë¥¼ ì…ë ¥ë°›ìŠµë‹ˆë‹¤
candis = ['ì§„êµí›ˆ', 'ê¹€íƒœìš°', 'ê¶Œìˆ˜ì •']

for candi in candis:
    search_query = f"ê°•ì„œêµ¬ì²­ì¥ {candi}"

    # YouTube ë™ì˜ìƒ ê²€ìƒ‰ì„ ìœ„í•œ ìš”ì²­ì„ ë§Œë“­ë‹ˆë‹¤
    search_request = youtube.search().list(
        q=search_query,
        type="video",
        part="id",
        maxResults=300
    )

    # API ìš”ì²­ì„ ì‹¤í–‰í•˜ê³  ì‘ë‹µì„ ë°›ìŠµë‹ˆë‹¤
    search_response = search_request.execute()

    # ê²€ìƒ‰ ê²°ê³¼ë¥¼ ë°ì´í„°í”„ë ˆì„ìœ¼ë¡œ ë³€í™˜í•©ë‹ˆë‹¤
    video_data = []
    for item in search_response["items"]:
        video_id = item["id"]["videoId"]

        # ë™ì˜ìƒ ì •ë³´ë¥¼ ê°€ì ¸ì˜¤ëŠ” ìš”ì²­ì„ ë§Œë“­ë‹ˆë‹¤
        video_request = youtube.videos().list(
            part="snippet,statistics",
            id=video_id
        )

        # API ìš”ì²­ì„ ì‹¤í–‰í•˜ê³  ì‘ë‹µì„ ë°›ìŠµë‹ˆë‹¤
        video_response = video_request.execute()

        # ê°€ì ¸ì˜¨ ë™ì˜ìƒ ì •ë³´ë¥¼ ë¦¬ìŠ¤íŠ¸ì— ì €ì¥í•©ë‹ˆë‹¤
        for video_item in video_response["items"]:
            video_title = video_item["snippet"]["title"]
            video_publish_date = video_item["snippet"]["publishedAt"]
            video_likes = video_item["statistics"]["likeCount"]

        # ë™ì˜ìƒì˜ ëŒ“ê¸€ì„ ê°€ì ¸ì˜¤ëŠ” ìš”ì²­ì„ ë§Œë“­ë‹ˆë‹¤
        comments_request = youtube.commentThreads().list(
            part="snippet",
            videoId=video_id,
            textFormat="plainText"
        )

        # API ìš”ì²­ì„ ì‹¤í–‰í•˜ê³  ì‘ë‹µì„ ë°›ìŠµë‹ˆë‹¤
        comments_response = comments_request.execute()

        # ê°€ì ¸ì˜¨ ëŒ“ê¸€ ì •ë³´ë¥¼ ë¦¬ìŠ¤íŠ¸ì— ì €ì¥í•©ë‹ˆë‹¤
        for comment_item in comments_response["items"]:
            comment_text = comment_item["snippet"]["topLevelComment"]["snippet"]["textDisplay"]
            comment_author = comment_item["snippet"]["topLevelComment"]["snippet"]["authorDisplayName"]
            comment_publish_date = comment_item["snippet"]["topLevelComment"]["snippet"]["publishedAt"]
            comment_likes = comment_item["snippet"]["topLevelComment"]["snippet"]["likeCount"]

            video_data.append([video_title, video_publish_date, video_likes, comment_text, comment_author, comment_publish_date, comment_likes])

    # ë°ì´í„°í”„ë ˆì„ì„ ìƒì„±í•©ë‹ˆë‹¤
    columns = ["ë™ì˜ìƒ ì œëª©", "ê²Œì‹œì¼", "ì˜ìƒ ì¢‹ì•„ìš” ìˆ˜", "ëŒ“ê¸€", "ì‘ì„±ì", "ëŒ“ê¸€ ì‘ì„±ì¼", "ëŒ“ê¸€ ì¢‹ì•„ìš” ìˆ˜"]
    merged_df = pd.DataFrame(video_data, columns=columns)

    # ë°ì´í„°í”„ë ˆì„ì„ CSV íŒŒì¼ë¡œ ì €ì¥í•©ë‹ˆë‹¤
    merged_df.to_csv(f"results_{candi}.csv", index=False, encoding="utf-8-sig")

    print("ê²€ìƒ‰ ê²°ê³¼ë¥¼ CSV íŒŒì¼ë¡œ ì €ì¥í–ˆìŠµë‹ˆë‹¤.")

    time.sleep(10)
```
